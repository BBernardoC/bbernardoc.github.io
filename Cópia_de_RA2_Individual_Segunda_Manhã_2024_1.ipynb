# Questão 3 - Vale até 4 Pontos

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a variável simbólica
x = sp.symbols('x')

# Definindo a função f(x)
f = sp.exp(2*x) - sp.exp(x)

# Calculando a derivada de f(x)
df = sp.diff(f, x)

# Encontrando os pontos críticos (onde a derivada é zero)
critical_points = sp.solve(df, x)

# Calculando o valor da função nos pontos críticos
values_at_critical_points = [f.subs(x, p) for p in critical_points]

# Mostrando os resultados
print("Pontos críticos:", critical_points)

# Plotando o gráfico da função
x_values = np.linspace(-2, 2, 400)
y_values = np.exp(2*x_values) - np.exp(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = e^{2x} - e^x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.axvline(critical_points, color='purple', linestyle='--', label='Ponto de Mínimo em $x=-log(2)$')



plt.grid(True)
plt.legend()
plt.show()
```

# Questão 4 - Alternativa - Vale até 2 pontos

Um pedaço reto de arame com $40 cm$ de comprimento será cortado em dois pedaços. Um pedaço será dobrado em um círculo e o outro será dobrado em um quadrado. Como o arame deve ser cortado para que a área total do círculo e do quadrado somadas seja a mínima?


primeiro definimos uma funcao que explica a relacao entre o arame e como sera cortado

$$ a+b=40$$
em que a é o circulo e b o quadrado


$$ \text{area do circulo} = πr^2$$
como o comprimento do arame do segmento do arame que será utilizado para fazer p circulo é igual ao comprimento da circunferencia temos que $y=2πr$ ou seja $r = \frac{y^2}{4π}$

$$ \text{area do quadrado} = b^2$$

substituindo temos que

$$f(x,y) = \frac{y^2}{4π}+b^2 $$

substituindo $y = 40-x$ temos

$$f(x,y) = \frac{(40-x)^2}{4π}+b^2 $$

agora se derivarmos em funcao de x, igualarmos igual a 0 e resolver em funcao de x , obteremos a area minima do quadrado o que nos leva ao resultado final

```python
import sympy as sp

# Definindo variáveis simbólicas
x = sp.Symbol('x')

# Definindo a função da área do círculo em termos de x
r = (40 - x) / (2 * sp.pi)
A_c = sp.pi * r**2

# Definindo a função da área do quadrado em termos de x
A_q = x**2

# Definindo a função da área total em termos de x
f = A_c + A_q

# Encontrando a derivada da função
f_prime = sp.diff(f, x)

# Encontrando os pontos críticos (onde a derivada é zero)
critical_points = sp.solve(f_prime, x)

# Filtrando os pontos críticos válidos (0 <= x <= 40)
valid_critical_points = [p for p in critical_points if 0 <= p <= 40]

# Calculando o valor da função nos pontos críticos válidos
min_area = min([f.subs(x, p) for p in valid_critical_points])

# Encontrando o comprimento do lado do quadrado correspondente ao ponto crítico que minimiza a área total
optimal_side_length = valid_critical_points[0]

# Calculando o raio correspondente ao ponto crítico que minimiza a área total
optimal_radius = (40 - optimal_side_length) / (2 * sp.pi)

# Somando os comprimentos dos segmentos de arame
total_wire_length = 2 * optimal_radius + optimal_side_length

# Imprimindo o resultado
print("Comprimento do lado do quadrado (x) que minimiza a área total:", optimal_side_length)
print("Comprimento do raio do círculo correspondente:", optimal_radius)
print("Comprimento total do arame:", total_wire_length)
print("Mínima área total:", min_area)
```
