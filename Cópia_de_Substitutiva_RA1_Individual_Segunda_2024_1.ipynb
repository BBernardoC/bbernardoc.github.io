```python
"""
EXERCICIO 1

"""


import sympy as sp
import numpy as np
from IPython.display import display, Math

# Definindo a variável simbólica e a expressão da função
x = sp.symbols('x')
expressao = ((x-1)*(x+1))/((x-1))

# Criando pontos próximos a 1 pela esquerda e pela direita, ponto de tendencia
valores_esquerda = np.linspace(1 - 0.001, 1, 10, endpoint=False)
valores_direita = np.linspace(1.0001, 1+ 0.001, 10, endpoint=False)

# Preparando cabeçalho da tabela em LaTeX
latex_str = r"\begin{array}{c|c|c|c}"
latex_str += r"\text{Ponto à esquerda} & \text{Valor da função} & \text{Ponto à direita} & \text{Valor da função} \\ \hline"

# Calculando os valores da expressão para os pontos dados e adicionando à string LaTeX
for esq, dir in zip(valores_esquerda, valores_direita):
    valor_esq = expressao.subs(x, esq).evalf()
    valor_dir = expressao.subs(x, dir).evalf()
    latex_str += f"{esq:.5f} & {valor_esq:.5f} & {dir:.5f} & {valor_dir:.5f} \\\\"

# Fechando a tabela em LaTeX
latex_str += r"\end{array}"

# Exibindo a tabela
display(Math(latex_str))
```

```python
"""
EXERCICIO 3

"""


import numpy as np
import sympy as sp
from IPython.display import display, Math

# Definindo as partes da função g(x)
x = sp.symbols('x')
piecewise_func = sp.Piecewise((x**2 + 1, x > 1), (2*x + 1, x <= 1))

# Criando os valores próximos a -3 para esquerda e direita
x_vals_left = np.linspace(0.9, 1, 10, endpoint=False)
x_vals_right = np.linspace(1, 1.1, 10, endpoint=False)

# Avaliando os valores da função
y_vals_left = [piecewise_func.subs(x, v).evalf() for v in x_vals_left]
y_vals_right = [piecewise_func.subs(x, v).evalf() for v in x_vals_right]

# Preparando a string LaTeX
latex_str = r"\begin{array}{c|c|c|c}"
latex_str += r"\text{Valor à esquerda de 1} & \text{Valor da função} & \text{Valor à direita de 1} & \text{Valor da função} \\ \hline"
for i in range(len(x_vals_left)):
    latex_str += r"{:.2f} & {:.2f} & {:.2f} & {:.2f} \\".format(x_vals_left[i], y_vals_left[i], x_vals_right[i], y_vals_right[i])
latex_str += r"\end{array}"

# Mostrando a tabela com notação matemática
display(Math(latex_str))
```

```python
"""

EXERCICIO 4

"""
# graficooo

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a variável simbólica
x = sp.symbols('x')

# Definindo a expressão da função
expressao = ((x * (x - 1))**1/2) - x


# Convertendo a expressão do SymPy para uma função NumPy
f = sp.lambdify(x, expressao, 'numpy')

# Gerando valores de x, evitando o ponto de descontinuidade em x=0
x_vals = np.linspace(-200, 200, 400)
x_vals = x_vals[x_vals != 0] # mudar essa linha != diferente do limite

# Calculando os valores da função
y_vals = f(x_vals)

# Plotando a função
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$(\sqrt{x^2 - x} - x)$')

# Mar(alterar para o limite de x)
x_point = float("inf")
y_point = f(x_point)
plt.scatter([x_point], [y_point], color='red')
plt.text(x_point, y_point, r'$\lim_{{x \to -\infty}} (\sqrt{{x^2 - x}} - x)$', verticalalignment='bottom', horizontalalignment='right')

# Configurações do gráfico
plt.title('Gráfico da função $(\sqrt{x^2 - x} - x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostrando o gráfico
plt.show()
```
