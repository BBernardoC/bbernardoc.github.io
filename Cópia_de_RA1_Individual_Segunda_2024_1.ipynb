1. **Primeira Questão:**

   $$\lim_{{x \to 0}} \frac{1 - \cos(x)}{x^2}$$


   Resposta:
 $$\lim_{{x \to 0}} \frac{1 - \cos(x)}{x^2} = lim_{{x \to 0}} \frac{{1 - \left(1 - \frac{{x^2}}{{2!}} + \frac{{x^4}}{{4!}} - \ldots \right)}}{{x^2}}$$


$$\frac{{\left(\frac{{x^2}}{{2!}} + \frac{{x^4}}{{4!}} - \ldots \right)}}{{x^2}}$$

$$\frac{{x^2 \left(\frac{1}{2} + \frac{x^2}{4}\right)}}{{x^2}}$$
$$\left(\frac{1}{2} + \frac{x^2}{4}\right)$$
$$\left(\frac{1}{2} + \frac{0^2}{4}\right) = \left(\frac{1}{2}\right)$$

Simplificando e aplicando os limites pode ser concluido que o resultado é 1/2, checando no gráfico podemos obter o mesmo resultado. Portanto pode se concluir que conforme x se aproxima de 0 o limite é 0.5

$$

$$ \lim_{{x \to 1}} \frac{{({x+8} - 9)}}{{(x-1) × (x+1) x (\sqrt{x+8}+3)}} =\lim_{{x \to 1}} \frac{{({x-1})}}{{(x-1) × (x+1) x (\sqrt{x+8}+3)}} $$

4. **Quarta Questão:**
  
  $$\lim_{t \to 0} \left( \frac{1}{t\sqrt{t + 1}} - \frac{1}{t} \right)$$



conforme pode ser visto no gráfico o limite é 1/2, pois conforme t se aproxima de 0 a funcao chega em -1/2


**Rúbrica de avaliação**: cada limite corretamente determinado com a algebra demonstrada em notação matemática 1 ponto; cada análise consistente 1,5 pontos. Total 2.5 pontos por questão implicando em, no máximo 10 pontos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(t):
    return (1 / (t * np.sqrt(t + 1))) - (np.sqrt(t + 1) / (t * np.sqrt(t + 1)))

# Criando um conjunto de valores t próximos de 0
t = np.linspace(-0.5, 0.5, 1000)  # Intervalo ampliado

# Calculando os valores correspondentes de f(t)
y = f(t)

# Plotando o gráfico da função
plt.plot(t, y, label=r'$\lim_{t \to 0} \left( \frac{1}{t\sqrt{t + 1}} - \frac{1}{t} \right)$')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Gráfico de $\\lim_{t \\to 0} \\left( \\frac{1}{t\\sqrt{t + 1}} - \\frac{1}{t} \\right)$')
plt.axhline(y=-0.5, color='r', linestyle='--', label='y = -0.5')
plt.axvline(x=0, color='k', linestyle='--', label='t = 0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return (1 - np.cos(x)) / (x**2)

# Gerando valores de x próximos a 0
x_values = np.linspace(-1, 1, 400)
x_values = x_values[x_values != 0]  # Removendo o ponto de singularidade

# Calculando os valores de f(x)
y_values = f(x_values)

# Traçando o gráfico
plt.plot(x_values, y_values, label='$\\frac{1 - \\cos(x)}{x^2}$', color='blue')
plt.xlabel('x')
plt.ylabel('$\\frac{1 - \\cos(x)}{x^2}$')
plt.title('Questão 1: Gráfico de $\\frac{1 - \\cos(x)}{x^2}$ quando $x \\rightarrow 0$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True)
plt.legend()
plt.show()
```
